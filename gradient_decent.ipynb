```python
import numpy as np
import matplotlib.pyplot as plt

# Data
np.random.seed(0)
X = np.linspace(1, 30, 30)
y = 1.5 * X + 3 + np.random.normal(0, 3, 30)  # Slope=1.5, Intercept=3, Noise~N(0,3)


# Initialize parameters
m = 0.0
b = 0.0

learning_rate = 0.01
epochs = 90

m_history = []
b_history = []

for _ in range(epochs):
    y_pred = m * X + b
    # Calculate gradients
    dm = (-2 / len(X)) * np.sum(X * (y - y_pred))
    db = (-2 / len(X)) * np.sum(y - y_pred)
    # Update parameters
    m -= learning_rate * dm
    b -= learning_rate * db
    # Store history
    m_history.append(m)
    b_history.append(b)

# Plot the line at each step
plt.scatter(X, y, color='red', label='Data points')
for i in range(0, epochs, 5):
    plt.plot(X, m_history[i] * X + b_history[i], label=f'Epoch {i+1}')
plt.plot(X, m_history[-1] * X + b_history[-1], color='green', linewidth=2, label='Final Fit')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Steps (Simple Linear Regression)')
plt.show()
```
